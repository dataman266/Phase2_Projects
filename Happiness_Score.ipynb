As it can be observed the Trust and Standard Error has outliers which might affect our analysis as well as the model, hence for these columns I will replace the outliers with the mean values.
So the outlier is any value that lies outside the range of Q1-1.5IQR to Q3+1.5IQR where Q1 is the first quartile, Q3 as third quarltile and IQR as interquartile range

```python
#List of columns with outliers
list=['Standard Error', 'Trust (Government Corruption)']
for col in list:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df.loc[(df[col] < lower_bound), col] = df[col].mean()
    df.loc[(df[col] > upper_bound), col] = df[col].mean()
```

```python
#Rechecking for outliers
#Checking for outliers
fig = plt.figure(figsize =(15, 10))
numerical_columns = ['Standard Error', 'Trust (Government Corruption)']
for i, col in enumerate(numerical_columns):
    plt.subplot(1, 2, i+1)
    plt.boxplot(df[col])
    plt.title(col)

plt.show()
```

As it can be observed that now the data the columns with outliers has only a few outliers hence, it won't affect a lot to out analysis as well as the model

# Now we will proceed with the building up of a model

```python
#Encoding the categorical column using one hot encoder as the Region columns has a multiple categorical values hence using dummies is the best option
df1 = pd.get_dummies(df, columns=['Region'])
df1
```

```python
# Define features and target variable
x = df1.drop(columns=["Country", "Happiness Rank", "Happiness Score"])
y = df1["Happiness Score"]
```

```python
#Splitting the data into training and testing dataset
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
#Importing the library for Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Initialize and train the linear regression model
lr= LinearRegression()
lr.fit(x_train, y_train)
lr
```

```python
# Predicting on the test set
y_pred = lr.predict(x_test)

# Evaluate the model's performance
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

mse, rmse, mae, r2
```

As the value of r2(R-squared) is very close to 1, it states that the model is performing very well. Alongwith it, the values of of mse(Mean_Sqaured_Error), rmse(Root_Mean_Sqaured_Error) and mae(Mean_Absolute_Error) being close to 0 supports the good performance and high accuracy of the model

```python
def happiness_scores(x):
    predictions = lr.predict(x)
    return predictions
test_predictions = happiness_scores(x_test)
print(test_predictions)
```

```python
#Now creating the function to predict the happiness for any arbitrary value of the variables under consideration
def score(model, standard_error, economy_gdp_per_capita, family, health_life_expectancy, freedom, trust_government_corruption, generosity, dystopia_residual, region):
    input_data = pd.DataFrame({
        'Standard Error': [standard_error],
        'Economy (GDP per Capita)': [economy_gdp_per_capita],
        'Family': [family],
        'Health (Life Expectancy)': [health_life_expectancy],
        'Freedom': [freedom],
        'Trust (Government Corruption)': [trust_government_corruption],
        'Generosity': [generosity],
        'Dystopia Residual': [dystopia_residual],
        'Region_Australia and New Zealand': [1 if region.lower() == 'australia and new zealand' else 0],
        'Region_Central and Eastern Europe': [1 if region.lower() == 'central and eastern europe' else 0],
        'Region_Eastern Asia': [1 if region.lower() == 'eastern asia' else 0],
        'Region_Latin America and Caribbean': [1 if region.lower() == 'latin america and caribbean' else 0],
        'Region_Middle East and Northern Africa': [1 if region.lower() == 'middle east and northern africa' else 0],
        'Region_North America': [1 if region.lower() == 'north america' else 0],
        'Region_Southeastern Asia': [1 if region.lower() == 'southeastern asia' else 0],
        'Region_Southern Asia': [1 if region.lower() == 'southern asia' else 0],
        'Region_Sub-Saharan Africa': [1 if region.lower() == 'sub-saharan africa' else 0],
        'Region_Western Europe': [1 if region.lower() == 'western europe' else 0]})
    predicted_score = model.predict(input_data)
    return predicted_score[0]
```

```python
predicted = score(lr, 0.034110, 1.39651, 1.34951, 0.94143, 0.66557, 0.143422, 0.29678, 2.51738, 'western europe')
print(f"Predicted happiness score", predicted)
```

As I predicted the happiness score for 1st row in the above example which is almost equal to the value present in the data, hence it can be said the model trained on the above data is performing well and eligible to find out the happiness score accurately on the basis of given input values.
